```python
!pip install selenium
!apt-get update # to update ubuntu to correctly run apt install
!apt install chromium-chromedriver
!cp /usr/lib/chromium-browser/chromedriver /usr/bin
!pip install beautifulsoup4
import sys
from bs4 import BeautifulSoup
import re

sys.path.insert(0,'/usr/lib/chromium-browser/chromedriver')
from selenium import webdriver
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
driver = webdriver.Chrome('chromedriver',chrome_options=chrome_options)
```

```python
#start with page number 1.

page=1;
seekUrl=f"https://www.seek.co.nz/jobs-in-information-communication-technology?page={page}&subclassification=6302%2C6290%2C6287%2C6286"

jobContentSeek=[]
numberOfJobs=1;
print('number of jobs',numberOfJobs)
while numberOfJobs !=0:
  seekUrl=f"https://www.seek.co.nz/jobs-in-information-communication-technology?page={page}&subclassification=6302%2C6290%2C6287%2C6286"
  driver.get(seekUrl)
  content = driver.page_source
  soup = BeautifulSoup(content)
  numberOfJobs=len(soup.findAll('article', attrs={'class':'_37iADb_'}))
  print('number of jobs',numberOfJobs)
  for a in soup.findAll('article', attrs={'class':'_37iADb_'}):
    jobTitle=a.find('a', attrs={'class':'_2S5REPk'})
    jobLink="https://www.seek.co.nz"+jobTitle.attrs['href']
    # print("jobs", jobLink)
    driver.get(jobLink)
    jobContent=driver.page_source
    jobSoup = BeautifulSoup(jobContent)
    for b in jobSoup.findAll('div', attrs={'class':'FYwKg d7v3r _1vNvq_4'}):
      jobPosition=b.find('h1', attrs={'class':'FYwKg C6ZIU_4 _3nVJR_4 _1H36Y_4 _2DNlq_4 _1NXQv_4'})
      jobDescription=b.find('div', attrs={'data-automation':'jobAdDetails'})
      jobLocation=b.find('div', attrs={'class':'FYwKg PrHFr _1EtT-_4'})
      jobType=b.find('div', attrs={'data-automation':'job-detail-work-type'})
      jobContentSeek.append({
          "name":jobTitle.text,
          "location":jobLocation.text,
          "link":jobLink,
          "jobType":jobType.text,
          "description":str(jobDescription)
      })
    # print("jobs", jobContentSeek)
    page=page+1


print('news',jobContentSeek)
```

```python
keywords=['angular','javascript','react','typescript','python','php','css','html','scss','database','Administration', 'oop']
keywordJobs={}
for keyw in keywords:
  countVar=0;
  for jo in jobContentSeek:
    if re.search(keyw, jo['description'], re.IGNORECASE):
    # if keyw in jo['description']:
      countVar=countVar+1
  print('number of jobs have keyword',keyw ,'is ',countVar)
  # print('finding',countVar)
```
